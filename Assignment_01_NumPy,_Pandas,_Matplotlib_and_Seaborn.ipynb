```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sl_df = Data[['Classification', 'SL']] # Create a new dataframe with Sepal Length (SL) for each Classification
sl_mean_df = sl_df.groupby('Classification').mean() # Group the data by species and calculate the mean Sepal Length
plt.bar(sl_mean_df.index, sl_mean_df['SL'])# Create a bar chart using Matplotlib 
plt.xlabel('Classification') # Add labels 
plt.ylabel('Sepal Length (cm)') # Add labels 
plt.title('Mean Sepal Length of Iris Classification') # Add title 
plt.show() # Show the plot
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Filter the SL values of the three Classification
setosa_sl = Data[Data['Classification'] == 'Iris-setosa']['SL']
versicolor_sl = Data[Data['Classification'] == 'Iris-versicolor']['SL']
virginica_sl = Data[Data['Classification'] == 'Iris-virginica']['SL']

# Create a list of the filtered SL values
data = [setosa_sl, versicolor_sl, virginica_sl]

# Set the figure size
plt.figure(figsize=(8, 6))

# Create a box plot
plt.boxplot(data, labels=['Iris-Setosa', 'Iris-Versicolor', 'Iris-Virginica'])

# Set the plot title and axis labels
plt.title('Box plot of Sepal Length by Classification')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')

# Show the plot
plt.show()
```
